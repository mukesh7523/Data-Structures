```python
# 2.  Explain the difference between mutable and immutable data types with examples.

# Ans :- Mutable Data Types
#     A mutable data type is one whose values can be changed after the object is created, without changing its memory address.

#     Immutable Data Types
#    An immutable data type is one whose values cannot be changed after the object is created.
#    If a modification is attempted, a new object is created with a different memory address.


# Mutable Data Type Example (List)

my_list = [1, 2, 3]
print(id(my_list))
my_list.append(4)
print(my_list)
print(id(my_list))

print("\n","Immutable Data Type Example \n")


#Immutable Data Type Example (String)

my_str = "Hello"
print(id(my_str))

my_str += " World"
print(my_str)
print(id(my_str))
```

```python
# 14.  How does the “in” keyword work differently for lists and dictionaries.

# Ans :- The in keyword is used to check membership in both lists and dictionaries, but it behaves differently in each case.

#   in with Lists-

#  -In a list, in checks if a value exists.
#  -The search is linear (O(n)), meaning it scans each element until it finds a match.

#   in with Dictionaries

#  -In a dictionary, in checks if a key exists, not a value.
#  -This lookup is very fast (O(1)) because dictionaries use hash tables.
```

```python
# 16. What is a nested dictionary, and give an example of its use case.

# Ans :- A nested dictionary is a dictionary inside another dictionary.
#        It allows hierarchical data storage, where each key can map to another dictionary instead of a single value.


#   Example: Using a Nested Dictionary-

students = {
    "101": {"name": "Alice", "age": 22, "grades": {"math": 85, "science": 90}},
    "102": {"name": "Bob", "age": 24, "grades": {"math": 78, "science": 88}},
    "103": {"name": "Charlie", "age": 21, "grades": {"math": 92, "science": 95}}
}


print(students["101"]["grades"]["science"])
```

```python
# 17. Describe the time complexity of accessing elements in a dictionary.

# Ans :- Python dictionaries are implemented using hash tables, allowing for fast key-based lookups.
#      The average-case time complexity for accessing elements is O(1) (constant time), but in rare cases, it can degrade to O(n) (linear time) due to hash collisions.

# Ex-

data = {"Alice": 25, "Bob": 30, "Charlie": 22}
print(data["Bob"])
```

```python
# 19. Why are dictionaries considered unordered, and how does that affect data retrieval.

# Ans :- Historically, dictionaries were considered unordered because they stored key-value pairs using a hash table, where elements were placed based on their hash values, not in the order they were inserted.

#   However, starting from Python 3.7, dictionaries preserve insertion order, meaning that when you iterate over a dictionary, the elements appear in the order they were added.

# -  No Positional Indexing Like Lists
#      Unlike lists, where you can access elements by position (list[0]), dictionaries require key-based access.

# - Iteration Order (Depends on Python Version
#    Before Python 3.7 → Dictionaries did not guarantee order.
#    Python 3.7+ → Insertion order is preserved when iterating.

# - Lookup Performance (O(1))
#     Since dictionaries use hashing, key lookups are O(1) (constant time), making them much faster than list searches (O(n)).



```

```python
# 19.  Write a code to find the union of two sets {1, 2, 3} and {3, 4, 5}

# Ans :-

set1 = {1, 2, 3}
set2 = {3, 4, 5}


union_set = set1.union(set2)
print("Union using union():", union_set)


union_set_alt = set1 | set2
print("Union using | operator:", union_set_alt)
```

```python
# 20.  Write a code to find the intersection of two sets {1, 2, 3} and {2, 3, 4}

# Ans :-

set1 = {1, 2, 3}
set2 = {2, 3, 4}


intersection_set = set1.intersection(set2)
print("Intersection using intersection():", intersection_set)


intersection_set_alt = set1 & set2
print("Intersection using & operator:", intersection_set_alt)
```

```python
# 29.  Write a code to combine two dictionaries into one and print the result.

# Ans :-



dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}


combined_dict = dict1.copy()
combined_dict.update(dict2)

print("Combined Dictionary (using update()):", combined_dict)


combined_dict_alt = dict1 | dict2
print("Combined Dictionary (using | operator):", combined_dict_alt)
```
